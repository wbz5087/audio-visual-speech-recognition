```python
import subprocess
import os
import sys
```

```python
def generate_file_list(path,p):
    print("[INFO] loading images...")
    data = []
    labels = []
    train_y = []
    # grab the image paths and randomly shuffle them
    
    labels = os.listdir(path)
    labels.remove('.DS_Store')
    imagePaths = []
    print(labels)
    for i, label in enumerate(labels):
        path_file=os.path.join(path, label)
        #path_file = os.path.join(path,'label')
        path_files = os.path.join(path_file,p)
        imagePath = os.listdir(path_files)
        #print(path_files)
        #imagePath.remove('.DS_Store')
        for img in imagePath:
            if img == '.DS_Store':
                print('wrong')
            else:
                image = os.path.join(path_files,img)
                if image.endswith('.mp4'):
                    imagePaths.append(image)
            #print(path_files)
    #print(imagePaths)
    #random.seed(42)
    #random.shuffle(imagePaths)
    
    return imagePaths
```

```python
part='train'
trainlist = generate_file_list(path='data/',p=part) 

fileObject = open('train_video.txt', 'w')
for ip in trainlist:
    fileObject.write(ip)
    fileObject.write('\n')
fileObject.close()
```

```python

# Pre...
#textfile_path = 'filelist.txt'
textfile_path = 'val_video.txt'
# Read the text file
with open(textfile_path) as f:
    content = f.readlines()
# you may also want to remove whitespace characters like n at the end of each line
files_list = [x.strip() for x in content]
# Extract audio from video.
# It already save the video file using the named defined by output_name.
for file_num, file_path_input in enumerate(files_list, start=1):
    # Get the file name withoutextension
    file_name = os.path.basename(file_path_input)
    if 'mouthcropped' not in file_name:
        raw_file_name = os.path.basename(file_name).split('.')[0]
        #print(raw_file_name)
        file_dir = os.path.dirname(file_path_input)
        #print(file_dir)
        file_dir = os.path.join('audio/',file_dir)
        folder = os.path.exists(file_dir)
        if not folder:
            os.makedirs(file_dir)
            #print("--- new folder ---")
        #print(file_dir)
        file_path_output = file_dir + '/' + raw_file_name + '.wav'
        print(file_path_output)
        print('processing file: %s' % file_path_input)
        subprocess.call(
            ['ffmpeg', '-i', file_path_input, '-f', 'wav', '-vn', file_path_output])
        print('file %s saved' % file_path_output)
```
